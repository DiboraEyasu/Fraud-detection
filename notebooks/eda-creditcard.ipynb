```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look consistent
sns.set(style="whitegrid")
```

Load the dataset

```python
data = pd.read_csv("../data/raw/creditcard.csv")
```

No missing values were found in the creditcard dataset as in the fraud dataset so no imputation, filling or dropping rows is needed.

Handle duplicated values by dropping

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
```

```python
print(f"The shape of the data after dropping duplicated rows is {data.shape}")
```

Univariate Analysis 

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Class', data=data)
plt.title("Class distribution(fraud Vs non fraud)")
plt.show()
```

```python
data['Class'].value_counts(normalize=True) * 100
```

Fraud transactions represent approximately 0.167 % of the dataset which shows severe class imbalance

```python
plt.figure(figsize=(6,4))
sns.histplot(data['Amount'], kde=True, bins=50)
plt.xscale('log')
plt.title("Distribution of transaction Amount (log_scale)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Time'], bins=30)
plt.title("Transaction Time Distribution")
plt.show()
```

Bivariate Analysis

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=data['Class'], y=data['Amount'], palette='magma')
plt.title("Transaction Amount by class")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Class', y='Time', data=data)
plt.title("Transaction Amount by class")
plt.show()
```

Transaction and time plotting shows us that fraudulent transactions often take place faster since the first transaction. 

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x='Class', y='V7', data=data)
plt.title("V7 feature random visualization")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x='Class', y='V13', data=data)
plt.title("V13 feature random visualization")
plt.show()
```
